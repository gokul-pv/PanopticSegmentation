**Panoptic file**

```python
# JSON file
f = open ('/content/drive/MyDrive/EVA6/Capstone/dataset/panoptic.json', "r")
 
# Reading from file
data = json.loads(f.read())
```

```python
data.keys()
```

```python
LICENSES = data['licenses']
LICENSES
```

```python
INFO = data['info']
INFO
```

```python
CATEGORIES = data['categories']
CATEGORIES
```

```python
image_data = data['images']
image_data[1]
```

```python
annotation_data = data['annotations']
annotation_data[1]
```

```python
import pandas as pd

df_image = pd.DataFrame(image_data)
df_image
```

```python
df_annotation = pd.DataFrame(annotation_data)
df_annotation
```

```python
df_result = pd.concat([df_image, df_annotation], axis=1)
df_result
```

```python
# Creating a dataframe with 75%
# values of original dataframe
df_train = df_result.sample(frac = 0.8,random_state=1)
  
# Creating dataframe with 
# rest of the 25% values
df_test = df_result.drop(df_train.index)
  
print("\n80% of the givem DataFrame:")
print(df_train)
  
print("\nrest 20% of the given DataFrame:")
print(df_test)
```

```python
train_id = df_train['id'].values.tolist()
tes_id   = df_test['id'].values.tolist()
```

```python
df_train_images = df_train.iloc[:,:-3].to_dict('index')
df_train_annotations = df_train.iloc[:,7:].to_dict('index')
```

```python
train_images = [j for i,j in df_train_images.items()]
train_annotations = [j for i,j in df_train_annotations.items()]
```

```python
df_test_images = df_test.iloc[:,:-3].to_dict('index')
df_test_annotations = df_test.iloc[:,7:].to_dict('index')
```

```python
test_images = [j for i,j in df_test_images.items()]
test_annotations = [j for i,j in df_test_annotations.items()]
```

```python
panoptic_train = {
    'licenses': LICENSES, 
    'info': INFO, 
    'categories': CATEGORIES,
    'images': train_images,
    'annotations': train_annotations
}
```

```python
panoptic_val = {
    'licenses': LICENSES, 
    'info': INFO, 
    'categories': CATEGORIES,
    'images': test_images,
    'annotations': test_annotations
}
```

```python
with open('/content/drive/MyDrive/EVA6/Capstone/dataset/panoptic_train.json', "w") as panoptic_train_file:
    json.dump(panoptic_train, panoptic_train_file)
    
with open('/content/drive/MyDrive/EVA6/Capstone/dataset/panoptic_val.json', "w") as panoptic_val_file:
    json.dump(panoptic_val, panoptic_val_file)
    
```

**COCO file**

```python
# JSON file
f1 = open ('/content/drive/MyDrive/EVA6/Capstone/dataset/coco.json', "r")
 
# Reading from file
coco_data = json.loads(f1.read())
```

```python
coco_data.keys()
```

```python
coco_train_images = []
coco_test_images = []
coco_train_annotation = []
coco_test_annotation = []
```

```python
for im in coco_data['images']:
  if (im['id'] in train_id):
    coco_train_images.append(im)
  else:
    coco_test_images.append(im)
```

```python
for ann in coco_data['annotations']:
  if ann['image_id'] in train_id:
    coco_train_annotation.append(ann)
  else:
    coco_test_annotation.append(ann)
```

```python
coco_train = {
    'licenses': coco_data['licenses'], 
    'info': coco_data['info'], 
    'categories': coco_data['categories'],
    'images': coco_train_images,
    'annotations': coco_train_annotation
}
```

```python
coco_val = {
    'licenses': coco_data['licenses'], 
    'info': coco_data['info'], 
    'categories': coco_data['categories'],
    'images': coco_test_images,
    'annotations': coco_test_annotation
}
```

```python
with open('/content/drive/MyDrive/EVA6/Capstone/dataset/coco_train.json', "w") as coco_train_file:
    json.dump(coco_train, coco_train_file)
    
with open('/content/drive/MyDrive/EVA6/Capstone/dataset/coco_val.json', "w") as coco_val_file:
    json.dump(coco_val, coco_val_file)
```
